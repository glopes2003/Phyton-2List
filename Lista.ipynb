```python
# 2. **Agenda de Contatos**: Desenvolva um programa que permita ao usuário adicionar, listar, buscar e remover contatos em uma agenda.
# Os contatos devem ser armazenados em um arquivo CSV.

agenda = {}

def add():
    nome = str(input("Nome: "))
    tel = str(input("Telefone: "))
    agenda[nome] = tel
    print("Contato adicionado com sucesso.")

def list():
    print("Contatos:")
    for nome, tel in agenda.items():
        print(f"Nome: {nome} Telefone: {tel}")

def search():
    nome = str(input("Nome do contato que deseja buscar: "))
    if nome in agenda:
        print(f"Nome: {nome} Telefone: {agenda[nome]}")
    else:
        print(f"Contato não encontrado :( ")

def remove():
    nome = str(input("Nome do contato que deseja remover: "))
    if nome in agenda:
        del agenda[nome]
        print(f"Contato removido com sucesso!")
    else:
        print(f"Contato não encontrado :(")

while True:
    print("\n<<< AGENDA >>>")
    print("1 - Adicionar Contato")
    print("2 - Listar Contatos")
    print("3 - Buscar Contato")
    print("4 - Remover Contato")

    opcao = str(input("Escolha uma opção: "))

    if opcao == '1':
        add()
    elif opcao == '2':
        list()
    elif opcao == '3':
        search()
    elif opcao == '4':
        remove()
        break
    else:
        print("Opção inválida! Por favor, escolha uma opção válida.")
```

```python
# 3. **Jogo da Forca**: Implemente o clássico jogo da forca em Python. O programa deve escolher uma palavra aleatória de uma lista
# e permitir que o jogador adivinhe as letras.


import random

# Lista de palavras para o jogo
palavras = ["arroz", "cachorro", "torta", "cartas"]

def random_word():
    return random.choice(palavras)

def play(palavra):
    palavra = palavra.lower()
    letras_corretas = set()
    letras_incorretas = set()

    while True:
        palavra_oculta = ""
        for letra in palavra:
            if letra in letras_corretas:
                palavra_oculta += letra
            else:
                palavra_oculta += "_"

        print(f"Palavra: {palavra_oculta}")

        letra_tentativa = input("\nDigite uma letra: ").lower()

        if len(letra_tentativa) != 1 or not letra_tentativa.isalpha():
            print("Por favor, insira uma única letra válida.")
            continue

        if letra_tentativa in letras_corretas or letra_tentativa in letras_incorretas:
            print("Você já tentou essa letra.")
        elif letra_tentativa in palavra:
            letras_corretas.add(letra_tentativa)
            if letras_corretas == set(palavra):
                print(f"Parabéns, você adivinhou a palavra!")
                print(f"Palavra: {palavra} ")
                break
        else:
            letras_incorretas.add(letra_tentativa)
            print("Letra incorreta. Tente novamente.")

def main():
    print("<<< Jogo da Forca >>>")
    palavra = random_word()
    play(palavra)

if __name__ == "__main__":
    main()
```

```python
# 4. **Simulador de Caixa Eletrônico**: Crie um programa que simule um caixa eletrônico. Ele deve permitir que os usuários consultem saldo, saquem dinheiro
# e depositem dinheiro em uma conta fictícia.

conta = {
  "nome": "Giovana",
  "amount": 0
}

def check():
  print(f"Seu saldo atual é de {conta['amount']} reais")

def deposit():
  nome = str(input("Usuário da Conta: "))
  if nome == conta["nome"] :
      amount = float(input("Insira o valor que deseja depositar: "))
      conta["amount"] += amount
      print(f"Depósito no valor de {amount} reais realizado com sucesso!")

def remove():
  nome = str(input("Usuário da Conta: "))
  if nome == conta["nome"] :
    amount = float(input("Insira o valor que deseja sacar: "))
    conta["amount"] -= amount
    print(f"Você acaba de realizar um saque de {amount} reais")

while True:
    print("\n<<< Banco >>>")
    print("1 - Checar Saldo")
    print("2 - Depositar")
    print("3 - Sacar")

    opcao = str(input("Escolha uma opção: "))

    if opcao == '1':
        check()
    elif opcao == '2':
        deposit()
    elif opcao == '3':
        remove()
    else:
        print("Opção inválida! Por favor, escolha uma opção válida.")
```

```python
# 6. **Gerenciador de Tarefas**: Crie um aplicativo de gerenciamento de tarefas que permita ao usuário adicionar, listar, marcar como concluída
# e remover tarefas. As tarefas devem ser armazenadas em um arquivo JSON.

tarefas = []

def add():
    descricao = input("Insira a tarefa: ")
    tarefas.append({"descricao": descricao, "concluida": False})
    print("Tarefa adicionada com sucesso!")

def list():
    if not tarefas:
        print("Nenhuma tarefa encontrada :( ")
    else:
        print("<<< Lista de tarefas >>>")
        for i, tarefa in enumerate(tarefas, 1):
            status = "Concluída" if tarefa["concluida"] else "Pendente"
            print(f"{i}. {tarefa['descricao']} - {status}")

def concluded():
    list()
    numero_tarefa = int(input("Digite o número da tarefa concluída: "))
    if 1 <= numero_tarefa <= len(tarefas):
        tarefas[numero_tarefa - 1]["concluida"] = True
        print("Tarefa marcada como concluída.")
    else:
        print("Número de tarefa inválido.")

def remove():
    list()
    numero_tarefa = int(input("Digite o número da tarefa a ser removida: "))
    if 1 <= numero_tarefa <= len(tarefas):
        tarefa_removida = tarefas.pop(numero_tarefa - 1)
        print(f"Tarefa removida: {tarefa_removida['descricao']}")
    else:
        print("Número de tarefa inválido.")

def main():
    while True:
        print("\n <<< Gerenciamento de Tarefas >>>")
        print("1 - Adicionar Tarefa")
        print("2 - Listar Tarefas")
        print("3 - Marcar Tarefa como Concluída")
        print("4 - Remover Tarefa")

        opcao = input("Escolha a opção: ")

        if opcao == "1":
            add()
        elif opcao == "2":
            list()
        elif opcao == "3":
            concluded()
        elif opcao == "4":
            remove()
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```

```python
# 7. **Adivinhe o Número**: Escreva um programa que gere um número aleatório e desafie o usuário a adivinhá-lo.
# O programa deve fornecer dicas (maior/menor) após cada tentativa.

import random

def guess():
    secretNumber = random.randint(1, 10)

    print("<<< Jogo de Adivinhação >>> ")
    print("Tente adivinhar o número entre 1 e 10")

    while True:
        attempt = 0
        attempt = int(input("Qual o número? "))
        attempt += 1

        if attempt < secretNumber:
            print("Tente novamente, o número secreto é maior")
        elif attempt > secretNumber:
            print("Tente novamente, o número secreto é menor")
        else:
            print(f"Parabéns, você adivinhou o número secreto!!!")
            break

if __name__ == "__main__":
    guess()
```

```python
#9. **Sistema de Reservas de Cinema**: Crie um sistema de reservas de cinema que permita aos usuários escolher filmes, horários e comprar ingressos.
#Os dados das reservas podem ser armazenados em um arquivo de texto.

def list():
    filmes = {
        "Lorax": ["10:00", "15:30", "19:00"],
        "Os Incríveis": ["11:30", "16:45", "20:15"],
        "Procurando Nemo": ["13:00", "18:00"]
    }

    print("\n<<< Filmes disponíveis >>>")
    for filme in filmes:
        print(filme)
    return filmes

def reservation():
    filmes = list()
    filme_escolhido = input("\nEscolha o filme: ")
    if filme_escolhido not in filmes:
        print("Infelizmente, o filme escolhido não foi encontrado.")
        return

    horarios = filmes[filme_escolhido]
    print(f"\nHorários disponíveis para {filme_escolhido}:")
    for horario in horarios:
        print(horario)

    horario_escolhido = input("\nEscolha o horário: ")
    if horario_escolhido not in horarios:
        print("Horário indisponível.")
        return

    qtd_ingressos = int(input("\nQuantos ingressos deseja comprar? "))

    with open("reservas.txt", "a") as arquivo_reservas:
        arquivo_reservas.write(f"Filme: {filme_escolhido}, Horário: {horario_escolhido}, Ingressos: {qtd_ingressos}\n")

    print("Reserva realizada com sucesso!")

def see_reservation():
    try:
        with open("reservas.txt", "r") as arquivo_reservas:
            reservas = arquivo_reservas.read()
            if not reservas:
                print("Nenhuma reserva encontrada.")
            else:
                print("\nReservas existentes:")
                print(reservas)
    except FileNotFoundError:
        print("Nenhuma reserva encontrada.")

def main():
    while True:
        print("\n <<< Reservas de Cinema >>>")
        print("1 - Listar Filmes")
        print("2 - Realizar Reserva")
        print("3 - Visualizar Reservas")

        opcao = input("Escolha a opção: ")

        if opcao == "1":
            list()
        elif opcao == "2":
            reservation()
        elif opcao == "3":
            see_reservation()
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
